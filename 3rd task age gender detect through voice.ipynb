```python
import numpy as np
import librosa
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv1D, MaxPooling1D
import tkinter as tk
from tkinter import filedialog, messagebox
import os

# Load or train models as required
def load_gender_age_model():
    # Create a simple CNN model for demonstration
    model = Sequential([
        Conv1D(32, 3, activation='relu', input_shape=(40, 1)),
        MaxPooling1D(2),
        Dropout(0.2),
        Conv1D(64, 3, activation='relu'),
        MaxPooling1D(2),
        Dropout(0.2),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(2, activation='softmax')  # Binary output for gender (male/female)
    ])
    # model.load_weights("path_to_gender_age_model.h5")  # Load trained weights here
    return model

def load_emotion_model():
    model = Sequential([
        Conv1D(32, 3, activation='relu', input_shape=(40, 1)),
        MaxPooling1D(2),
        Dropout(0.2),
        Conv1D(64, 3, activation='relu'),
        MaxPooling1D(2),
        Dropout(0.2),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(4, activation='softmax')  # 4 emotions as an example
    ])
    # model.load_weights("path_to_emotion_model.h5")  # Load trained weights here
    return model

gender_age_model = load_gender_age_model()
emotion_model = load_emotion_model()

# Feature extraction function
def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=None)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
    mfccs = np.mean(mfccs.T, axis=0)
    return mfccs.reshape(40, 1)

# Process and classify the audio file
def process_audio(file_path):
    features = extract_features(file_path)

    # Predict gender
    gender_prediction = gender_age_model.predict(np.array([features]))
    gender = np.argmax(gender_prediction)  # Assuming 0 for female, 1 for male

    if gender == 0:
        messagebox.showinfo("Result", "Upload male voice.")
        return
    
    # Predict age (dummy age prediction for illustration)
    predicted_age = np.random.randint(20, 80)  # Replace with actual age prediction
    
    if predicted_age > 60:
        emotion_prediction = emotion_model.predict(np.array([features]))
        emotion = np.argmax(emotion_prediction)  # Assume 0-3 for emotions
        emotions = ['Happy', 'Sad', 'Angry', 'Neutral']
        message = f"Age: {predicted_age} (Senior Citizen)\nEmotion: {emotions[emotion]}"
    else:
        message = f"Age: {predicted_age}"

    messagebox.showinfo("Result", message)

# GUI with Tkinter
def open_file():
    file_path = filedialog.askopenfilename(filetypes=[("Audio Files", "*.wav *.mp3")])
    if file_path:
        process_audio(file_path)

app = tk.Tk()
app.title("Age and Emotion Detection")
app.geometry("300x200")

label = tk.Label(app, text="Upload a voice note to detect age and emotion")
label.pack(pady=20)

upload_button = tk.Button(app, text="Upload Voice Note", command=open_file)
upload_button.pack(pady=10)

app.mainloop()
```
